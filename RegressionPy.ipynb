# Regression in Python


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
import seaborn as sns
%matplotlib inline
import  statsmodels.api as sm
import statsmodels.stats.api as sms
from scipy.stats import boxcox
```

```python
manatees = pd.read_csv('manatees.csv')
```

```python
# plot the data
sns.pairplot(manatees)
```

```python
#histogram with ND curve for boats
sns.distplot(manatees['PowerBoats'])
```

```python
#histogram with ND curve for manatee deaths
sns.distplot(manatees['ManateeDeaths'])
```

```python
#next, test for homoscedasticity by first assigning variables to x and y
x = manatees['PowerBoats']
y = manatees['ManateeDeaths']
```

```python
#build the model
model = sm.OLS(y,x).fit()
```

```python
#calculate the residuals
pred_val = model.fittedvalues.copy()
true_val = manatees['ManateeDeaths'].values.copy()
residual = true_val - pred_val
```

```python
#graph the residuals
fig, ax = plt.subplots(figsize=(6, 2.5))
_ = ax.scatter(residual, pred_val)
```

```python
# check the graph with BP stat
sms.diagnostic.het_breuschpagan(residual, manatees[['PowerBoats']])
```

```python
# check the graph with Harvey Collier stat
sms.linear_harvey_collier(model)
```

```python
#transform the data to become more normalized using Box-cox
transformed, _ = boxcox(manatees['PowerBoats'])
```

```python
#plot the transformed data
plt.hist(transformed)
#we find that the heteroscedasticity has still not been corrected to homoscedasticity
```

```python
#next, pull these values into a model and rerun the tests for homoscedasticity by reassigning x as transformed instead of as powerboats
x = transformed
model1 = sm.OLS(y,x).fit()
model1.summary()
```

```python
#then create new residual values
pred_val = model1.fittedvalues.copy()
true_val = manatees['ManateeDeaths'].values.copy()
residual = true_val - pred_val
```

```python
#re plot
fig, ax = plt.subplots(figsize=(6, 2.5))
_ = ax.scatter(residual, pred_val)
```

```python
#multicolliinearity - done if you have more than 1 IV
#correlation matrix
manatees.corr()
```

```python
#heat map
sns.heatmap(manatees.corr(), annot=True)
```

### Screening for Outliers

```python
#create a plot that will test for influential data points (x and y space) by looking at leverage & studentized
#deleted residuals at the same time

fig, ax = plt.subplots(figsize=(12,8))
fig = sm.graphics.influence_plot(model, alpha = .05, ax = ax, criterion="cooks")

#observation: no studentized residuals greater than 2.5-3, and no leverage vals >.5, so good to go.
```

```python
#to view the influence info in table format instead of graphically:
infl = model.get_influence()
print(infl.summary_frame())
```
