```python
# load pkgs
import requests
from bs4 import BeautifulSoup
import nltk
from nltk.tokenize import RegexpTokenizer
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# read in text
url = 'https://www.gutenberg.org/files/11/11-h/11-h.htm'
```

```python
# make a request to get data from the webpage
r = requests.get(url)
```

```python
# Convert text to soup
html = r.text
soup = BeautifulSoup(html,"html.parser")
type(soup)
```

```python
# tokenize the data
text = soup.get_text()
tokenizer = RegexpTokenizer('\w+')
tokens = tokenizer.tokenize(text)
tokens[:5]
```

```python
# remove all capitalization
words = []
for word in tokens:
    words.append(word.lower())
```

```python
# remove stopwords
stopwords = nltk.corpus.stopwords.words('english')
```

```python
# filter out these stop words
wordsWithoutStops = []
for word in words:
    if word not in stopwords:
        wordsWithoutStops.append(word)
wordsWithoutStops[:5]
```

```python
# count and plot words
sns.set()
frequencyDis = nltk.FreqDist(wordsWithoutStops)
frequencyDis.plot(25)
```
