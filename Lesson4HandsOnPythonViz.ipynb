# Quantitative Data Visualization

```python
import pandas as pd
import seaborn as sns
import matplotlib as mlab
import matplotlib.pyplot as plt
```

### Part 1: Create a Histogram with 7 bars 

```python
powerBoats = pd.read_csv("L3P1.csv")
powerBoats
```

```python
print(powerBoats.columns)
```

```python
num_bins = 7
n, bins, patches = plt.hist(powerBoats["Power Boats (in 1000's)"], num_bins, facecolor='blue', alpha=1)
plt.xlabel("Power Boats Registered x 1000")
plt.title("Histogram of Power Boats Registered in Florida Over 35 Years")
```

### Part 2: Create a Bar Graph

```python
Cars = pd.read_csv("L3P2.csv")
Cars
```

```python
barChart = Cars['Cars'].value_counts().plot(kind='bar')
barChart.set_title("Frequency of Car Model")
barChart.set_xlabel("Model")
barChart.set_ylabel("Frequency")
```

### Part 3: Create a Stacked Bar Chart

```python
CarsLoc = pd.read_csv("L3Part3.csv")
CarsLoc
```

```python
crosstab_CarsLoc = pd.crosstab(CarsLoc['Car'], CarsLoc['Location'])
crosstab_CarsLoc.plot.bar(stacked=True)
```

### Part 4: Create a Scatterplot

```python
Croc = pd.read_csv("crocodiles.csv")
Croc
```

```python
CrocPlot = Croc.plot.scatter(x = 'HeadLength', y = 'BodyLength')
CrocPlot.set_title("Estuarine Crocodiles, Head Length and Body Length")
```

### Create a Line Chart

```python
Heart = pd.read_csv("L6handson.csv")
Heart
```

```python
# Combine Month, Day, and Year columns into one column named 'Date'
Heart['Date'] = Heart["Month"] + " " + Heart["Day"] + " " + Heart["Year"].map(str)
Heart
```

```python
plt.plot(Heart['Date'], Heart['Heart Attacks'])
plt.xlabel('Date')
plt.ylabel('Heart Attacks')
plt.title("Heart Attacks Over Time")
```
