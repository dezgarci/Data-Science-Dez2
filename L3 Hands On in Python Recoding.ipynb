# Lesson 3 Hands On Recoding in Python

```python
import pandas as pd
```

```python
eatingHabits = pd.read_csv("Eating_Habits.csv")
```

```python
eatingHabits.head()
```

### recode Activity into a new variable where 0= not eating chocolate, and 1 = eating chocolate

```python
eatingHabits.Activity.value_counts()
```

```python
def activity (series):
    if series == "Eating fruit":
        return 0
    if series == "Eating candy, chocolate bars":
        return 1
    if series == "Eating french fries":
        return 0
    if series == "Eating hamburgers, hot dogs, or sausages":
        return 0
    if series == "Eating whole wheat or rye bread":
        return 0
    if series == "Drinking soft drinks, cola or other drinks with sugar":
        return 0
    if series == "Drinking coffee":
        return 0
    if series == "Eating raw vegetables":
        return 0
    if series == "Eating potato chips, crisps":
        return 0
    if series == "Drinking low fat milk":
        return 0
    if series == "Drinking while milk":
        return 0
    if series == "Eating peanuts":
        return 0
    
eatingHabits['ChocolateR'] = eatingHabits['Activity'].apply(activity)
eatingHabits.ChocolateR.value_counts()
```

### recode frequency from texts to numbers starting at 0

```python
eatingHabits.Frequency.value_counts()
cleanFreq = {"Frequency" : {"Never" : 0, "Seldom" : 1, "At least once a week" : 2, "Once a day" : 3, "More than once a day" : 4}}
eatingHabits.replace(cleanFreq, inplace=True)
eatingHabits.head()
```

### recode sex from text to numbers starting at 0

```python
eatingHabits.Sex.value_counts()
cleanSx = {"Sex" : {"Females" : 0, "Males" : 1}}
eatingHabits.replace(cleanSx, inplace=True)
eatingHabits.head()
```

### Dummy code the Age group variable

```python
AgeDummy = pd.get_dummies(eatingHabits['Age group'], drop_first=True)
AgeDummy
```

```python
EatingHabits2 = pd.concat([eatingHabits, AgeDummy], axis=1)
EatingHabits2.head()
```
